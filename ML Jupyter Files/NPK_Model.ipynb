#REMOVE OUTLIERS FROM DATASET

```python
df_boston = data
df_boston.columns = df_boston.columns
df_boston.head()

print("old shape", df_boston.shape)

import numpy as np
import pandas as pd

'''file_path = 'Crop_recommendation_data_2.csv'
data = pd.read_csv(file_path)'''

columns = ['N', 'P', 'K', 'humidity', 'temperature']

outliers_indices = set()

for col in columns:
    Q1 = np.percentile(df_boston[col], 25, interpolation='midpoint')
    Q3 = np.percentile(df_boston[col], 75, interpolation='midpoint')
    IQR = Q3 - Q1

    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    upper_outliers = np.where(df_boston[col] >= upper_bound)[0]
    lower_outliers = np.where(df_boston[col] <= lower_bound)[0]

    outliers_indices.update(upper_outliers)
    outliers_indices.update(lower_outliers)

df_boston.drop(list(outliers_indices), inplace=True)

print("New shape:", data.shape)
```

```python
X = data.drop('label', axis=1)
y = data['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,shuffle=True, random_state=0)
```

#LGBM CLASSIFIER

```python
import lightgbm as lgb
model_lgb = lgb.LGBMClassifier()
model_lgb.fit(X_train, y_train)
y_pred = model_lgb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_pred, y_test)
print("lgbm model accuracy score of:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(20,20))
sns.heatmap(cm, annot=True, fmt=".0f", linewidths=0.5,square=True,cmap='Greens')
plt.ylabel("actual label")
plt.xlabel('predicted label')
all_smaple_title = 'Confusion Matrix - score:'+str(accuracy_score(y_test,y_pred))
plt.title(all_smaple_title, size =15)
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

#DESICION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
Classifier_dtr = DecisionTreeClassifier(criterion='entropy', random_state=0)
Classifier_dtr.fit(X_train,y_train)
y_pred1 = Classifier_dtr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_dtr = accuracy_score(y_pred1, y_test)
print("dtr model accuracy score of:", accuracy_dtr)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred1)

plt.figure(figsize=(20,20))
sns.heatmap(cm, annot=True, fmt=".0f", linewidths=0.5,square=True,cmap='Greens')
plt.ylabel("actual label")
plt.xlabel('predicted label')
all_smaple_title = 'Confusion Matrix - score:'+str(accuracy_score(y_test,y_pred1))
plt.title(all_smaple_title, size =15)
plt.show()
```

#RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
Classifier_rfr = RandomForestClassifier(criterion='entropy', random_state=0)
Classifier_rfr.fit(X_train,y_train)
y_pred2 = Classifier_rfr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_rand = accuracy_score(y_pred2, y_test)
print("randfor model accuracy score of:", accuracy_rand)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred2)

plt.figure(figsize=(20,20))
sns.heatmap(cm, annot=True, fmt=".0f", linewidths=0.5,square=True,cmap='Greens')
plt.ylabel("actual label")
plt.xlabel('predicted label')
all_smaple_title = 'Confusion Matrix - score:'+str(accuracy_score(y_test,y_pred2))
plt.title(all_smaple_title, size =15)
plt.show()
```

```python
def recommendations(N,P,K,temperature,humidity,ph):
    features = np.array([[N,P,K,temperature,humidity,ph]])
    prediction = Classifier_rfr.predict(features).reshape(1,-1)
    return prediction[0]
```

```python
N = 80
P = 90
K = 47
temperature = 26.59743595
humidity = 79.35898915
ph = 6.21084479

predict = recommendations(N,P,K,temperature,humidity,ph)
print(predict)
```
